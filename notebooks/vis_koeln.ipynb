# Visualize Landprices


```python
import pandas as pd
import geopandas as gpd
```

```python

# Read data
## Prices & Neighborhood
koeln_prices = pd.read_csv("../data/raw/1 Land Prices/Land_Prices_Neighborhood_Köln.csv", sep=";", encoding= "utf-8-sig")
koeln_gpd = gpd.read_file("../data/raw/3 Neighborhoods/Neighborhoods_Köln.gpkg")

koeln_buildings = pd.read_csv("../data/raw/2 Zensus/Zensus_Köln_Buildings.csv", sep=";", encoding= "utf-8-sig")
koeln_families = pd.read_csv("../data/raw/2 Zensus/Zensus_Köln_Families.csv", sep=";", encoding= "utf-8-sig")
koeln_households = pd.read_csv("../data/raw/2 Zensus/Zensus_Köln_Households.csv", sep=";", encoding= "utf-8-sig")
koeln_pop = pd.read_csv("../data/raw/2 Zensus/Zensus_Köln_Population.csv", sep=";", encoding= "utf-8-sig")
```

```python
koeln_prices
```

```python
koeln_gpd = koeln_gpd.merge(koeln_prices, on = "Neighborhood_FID", how = "inner")
```

```python
koeln_gpd.explore(name="Land_Value", column = "Land_Value", scheme = "quantiles", cmap = "RdYlBu_r", legend = True)
```
